```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("train_titanic.csv")
df.drop(["PassengerId","Name","Ticket","Cabin"],axis=1,inplace=True)
```

```python
df.fillna({"Age":df["Age"].median()},inplace=True)
df.fillna({"Embarked":df["Embarked"].mode()[0]},inplace=True)
df.isnull().values.any()
```

```python
label = LabelEncoder()
df["Sex"] = label.fit_transform(df["Sex"])
df["Embarked"] = label.fit_transform(df["Embarked"])
```

```python
x = df.drop("Survived", axis=1)
y = df["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model1 = LogisticRegression(max_iter=1000)
model1.fit(x_train_scaled, y_train)

y_pred1 = model1.predict(x_test_scaled)
accuracy1 = accuracy_score(y_test, y_pred1)
print(accuracy1*100,"%")
```

```python


knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_scaled, y_train)

y_pred2 = knn.predict(x_test_scaled)
accuracy2 = accuracy_score(y_test, y_pred2)
print(accuracy2*100,"%")
```

```python
model3= DecisionTreeClassifier(max_depth=5,min_samples_split=10,random_state=42)
model3.fit(x_train_scaled, y_train)

y_pred3 = model3.predict(x_test_scaled)
accuracy3 = accuracy_score(y_test, y_pred3)
print(accuracy3*100,"%")
```

```python
results = pd.DataFrame({"Model": ["Logistic Regression","KNN","Decision Tree"],"Accuracy": [accuracy1,accuracy2,accuracy3]})
print(results)
```
